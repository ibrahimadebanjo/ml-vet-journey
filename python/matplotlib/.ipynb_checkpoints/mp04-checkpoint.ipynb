```python
import numpy as np
import matplotlib.pyplot as plt
ages = np.random.normal(20,2.5,206)
# Histogram
plt.hist(ages, cumulative=True)
# attr bins= 20,
plt.show()
# piechart
lang = ["Python", "C++", "Java", "C#", "Go"]
votes = [50,24,14,6,17] #the numbers must aggregate to 100
explodes = [0.4,0, 0, 0,0.1]
plt.pie(votes, labels =lang, explode = explodes,autopct= "%.2f%%")
# Atrr autopct= "%2f%%",pctdistance=1.8,startangle=90
# Explodes is used to seperate specified value for each element in a piechart 
plt.show()
```
